### *Crea almenos una visualizacion para:*

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("DelayedFlights.csv") # cargo el Data Frame
df.head()
```

#### *Una variable categorica (UniqueCarrier)*

```python
ax1 = sns.histplot(data=df, x="UniqueCarrier")
```

### *Una variable numerica (ArrDelay)*

```python
ax2 = sns.displot(data=df, x="ArrDelay")
```

### *Una variable numerica y una categorica (ArrDelay y UniqueCarrier)*

```python
ax3 = sns.barplot(data=df, x="UniqueCarrier", y="ArrDelay")
```

### *Dos variables numericas (ArrDelay y DepDelay)*

```python
ax4 = sns.regplot(data=df,x="ArrDelay", y="DepDelay", line_kws={"color":"r","alpha":0.5,"lw":5})
```

### *Tres variables (ArrDelay, DepDelay y UniqueCarrier)*

```python
ax5 = sns.FacetGrid(data=df, col="UniqueCarrier", hue="UniqueCarrier", col_wrap=4)
ax5.map(sns.scatterplot, "ArrDelay", "DepDelay")
```

### *Mas de tres variables (ArrDelay, DepDelay, AirTime i UniqueCarrier)*

```python
fig = plt.figure(figsize=(15,9))
ax6 = fig.add_subplot(111, projection='3d')
for s in df.UniqueCarrier.unique():
    ax6.scatter(df.AirTime[df.UniqueCarrier==s],
                df.DepDelay[df.UniqueCarrier==s],
                df.ArrDelay[df.UniqueCarrier==s],
                label=s)
plt.legend(loc='upper left')
```

### *Velocidad media de vuelo*

```python
df['AvgSpeed'] = round(df['Distance']/(df['AirTime']/60)) # velocidad media
df.head()
```

```python
df.groupby(by="UniqueCarrier")["AvgSpeed"].mean() # agrupo por aerolineas 
```

```python
plt.figure(figsize=(10,3))
sns.boxplot(data=df, x='UniqueCarrier',y='AvgSpeed')
```

### *Si los vuelos han llegado tarde o no (True/False)*

```python
df['ArrLate'] = df['ArrDelay']>0 # si ha llegado tarde o no
df.head()
```

```python
df["ArrLate"].value_counts().plot.pie(autopct='%1.1f%%',shadow=True, startangle=90)
my_circle=plt.Circle( (0,0), 0.5, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)
```

### *Top 10 vuelos mas largos*

```python
df2 = df[["UniqueCarrier","AirTime"]] # me quedo con las dos columnas que me interesan
df2 = df2.sort_values(by="AirTime", ascending=False) # ordeno los valores de forma descendente
df2 = df2.head(10) # me quedo con los diez primeros
df2
```

```python
height = df2["AirTime"]
bars= df2["UniqueCarrier"]
y_pos = np.arange(len(bars))
plt.barh(y_pos, height)
plt.yticks(y_pos, bars)
plt.show()
```

### *Top 10 vuelos mas retrasados*

```python
df3 = df[["UniqueCarrier","ArrDelay"]] # me quedo con las dos columnas que me interesan
df3 = df3.sort_values(by="ArrDelay", ascending=False) # ordeno los valores de manera descendente
df3 = df3.head(10) # me quedo con los 10 primeros
df3
```

```python
height = df3["ArrDelay"]
bars = df3["UniqueCarrier"]
x_pos = np.arange(len(bars))
plt.bar(x_pos, height, color=(0.1, 0.1, 0.1, 0.1),  edgecolor='blue')
plt.xticks(x_pos, bars)
plt.show()
```
